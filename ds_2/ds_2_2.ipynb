```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### 1. Read the 2017_jun_final.csv file using the `read_csv` method

```python
df = pd.read_csv("2017_jun_final.csv")
```

#### 2. Read the resulting table using the `head` method

```python
df.head()
```

```python
# rename column names
df = df.rename(columns={"Должность": "Position", 
                        "Язык.программирования": "Programming_language",
                        "Специализация": "Specialty",
                        "Общий.опыт.работы": "General_work_experience", 
                        "Опыт.работы.на.текущем.месте": "Work_experience_in_the_current_position",
                        "Зарплата.в.месяц": "Salary_per_month", 
                        "Изменение.зарплаты.за.12.месяцев": "Salary_change_for_12_months",
                        "Город": "City", 
                        "Размер.компании": "Company_size", 
                        "Тип.компании": "Company_type",
                        "Пол": "Sex", 
                        "Возраст": "Age", 
                        "Образование": "Education",
                        "Университет":"University",
                        "Еще.студент": "Student_now",
                        "Уровень.английского": "English_level", 
                        "Предметная.область": "Subject",
                        "Дата.заполнения": "Date_of_completion",
                        "Валюта": "Currency",
                        "User.Agent": "User_agent"})
```

#### 6. Remove all columns with blanks except for the "Мова програмування" column

```python
df.drop(["Specialty", "University", "Subject", "Currency", "cls"], axis=1, inplace=True)
```

```python
df
```

#### 7. Again, calculate the proportion of blanks in each column and make sure that only the "Programming_language" column remains

```python
df.isnull().sum()
```

#### 8. Delete all rows in the original table using the dropna method

```python
df = df.dropna()
```

#### 9. Define a new table size using the shape method

```python
df.shape
```

#### 10. Create a new table python_data, which will contain only rows with specialists who have specified the Python programming language

```python
python_lang = df.query('Programming_language == "Python"')
```

```python
python_lang.head()
```

#### 11. Resize the python_data table using the shape method

```python
python_lang.shape
```

#### 12. Using the groupby method, group by the "Position" column

```python
pos_group = python_lang.groupby(["Position"])
```

```python
pos_group.head()
```

#### 13. Create a new DataFrame, where for the grouped data by the "Position" column, perform data aggregation using the `agg` method and find the minimum and maximum values in the "Salary.per.month" column

```python
df_salary = pos_group["Salary_per_month"].agg(["min", "max"])
df_salary
```

#### 14. Create a `fill_avg_salary` function that returns the average salary per month. Use it for the `apply` method and create a new column "avg"

```python
def fill_avg_salary(data):
    return (data["min"]+data["max"])/2

avg = df_salary.apply(fill_avg_salary, axis=1, result_type='expand')
avg.name = "avg"
df_salary = pd.concat([df_salary, avg], axis=1)
df_salary
```

#### 15. Create descriptive statistics using the `describe` method for the new column.

```python
df_salary["avg"].describe()
```

#### 16. Save the resulting table to a CSV file

```python
df_salary.to_csv("Salary_per_month.csv")
```

## Visualization

```python
fig, ax = plt.subplots(figsize=(16, 14))
labels = [
    "Junior Software Engineer",
    "Senior Software Engineer",
    "Software Engineer",
    "System Architect",
    "Technical Lead",
]

data = [675, 5250, 2950, 4000, 3812.5]
explode = [0, 0.15, 0, 0, 0]
plt.pie(
    data,
    labels=labels,
    shadow=True,
    explode=explode,
    autopct="%.2f%%",
    pctdistance=1.15,
    labeldistance=1.25,
)

plt.title("Average salary per month (Python) in %", fontsize=16, loc='left')
plt.show()
```
