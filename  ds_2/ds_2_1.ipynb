Read the data using the `read_html` method from the table "Коефіцієнт народжуваності в регіонах України (1950-2019)" - [link](https://uk.m.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C)

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
url = "https://uk.m.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C"
df = pd.read_html(url,
                  attrs={"class": "wikitable collapsible collapsed", 
                         "style": "text-align: right"}, 
                  match="Регіон",
                  decimal=',', 
                  thousands='.')
df = df[0]
df
```

```python
df.shape
```

#### 3. Replace the value "-" in the table with the value `NaN`

```python
df = df.replace({"—": np.nan})
df.head()
```

#### 4. Determine the types of all columns using `dataframe.dtypes`

```python
df.dtypes
```

#### 5. Change the types of non-numeric columns to numeric. The hint is the columns where the symbol "-" was located

```python
df['2014'] = pd.to_numeric(df['2014'])
df['2019'] = pd.to_numeric(df['2019'])
df.dtypes
```

#### 6. Calculate the percentage of blanks in each column (use the `isnull` and `sum` methods)

```python
df.isnull().sum()/df.shape[0]
```

#### 7. Remove the data for the whole country from the table, the last row of the table

```python
df.drop(axis=0, index=27, inplace=True)
```

```python
df.tail()
```

```python
df
```

#### 8. Replace missing data in columns with the average values of these columns (`fillna` method)

```python
df = df.fillna({"1950": df["1950"].mean(), "1960": df["1960"].mean(), "1970": df["1970"].mean(),
                "1990": df["1990"].mean(), "2000": df["2000"].mean(), "2012": df["2012"].mean(),
                "2014": df["2014"].mean(), "2019": df["2019"].mean()})
df
```

#### 9. Get a list of regions where the birth rate in 2019 was higher than the average for Ukraine

```python
df[df["2019"] > df["2019"].mean()][["Регіон", "2019"]].sort_values("2019")
```

#### 10. Which region had the highest birth rate in 2014?

```python
df.sort_values(["2014"], ascending=False).head(1)[["Регіон", "2014"]]
```

#### 11. Build a bar chart of the birth rate by region in 2019

```python
fig, ax = plt.subplots(figsize=(16, 10))
df.plot(x="Регіон", y="2019", kind="bar", fontsize=10, color='g', ax=ax, grid=True)
plt.title("Birth rate by region in 2019", fontdict = {'fontsize' : 16})
ax.set_xlabel("Region", size=12, color="midnightblue")
ax.set_ylabel("Birth rate", size=12, color="midnightblue")
plt.show()
```

```python
df3 = pd.melt(df, id_vars=["Регіон"], 
                  var_name="Year", value_name="Rate")
df3 = df3.rename(columns={"Регіон": "Region"}) 

sns.set_theme(style="dark")

# Plot each region's time series in its own facet
g = sns.relplot(
    data=df3,
    x="Year", y="Rate", col="Region", hue="Region",
    kind="line", palette="crest", linewidth=4, zorder=5,
    col_wrap=3, height=2, aspect=1.5, legend=False,
)

# Iterate over each subplot to customize further
for year, ax in g.axes_dict.items():

    # Add the title as an annotation within the plot
    ax.text(.8, .85, year, transform=ax.transAxes, fontweight="bold")

    # Plot every year's time series in the background
    sns.lineplot(
        data=df3, x="Year", y="Rate", units="Region",
        estimator=None, color=".7", linewidth=1, ax=ax,
    )

# Reduce the frequency of the x axis ticks
ax.set_xticks(ax.get_xticks()[1::2])

# Tweak the supporting aspects of the plot
g.set_titles("")
g.set_axis_labels("", "Rate")
g.tight_layout()
```
